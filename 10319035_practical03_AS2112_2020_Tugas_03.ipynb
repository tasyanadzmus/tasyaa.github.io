```
#import modul yang diperlukan 

import numpy as np
import scipy.interpolate
import pylab
```

```
#Input data dapat ditanam di kode python, atau interaktif, atau baca dari file.
#Berikut ini merupakan contoh dari pembacaan file worksheet dalam format file teks CSV
#yang berisi dua kolom data dengan memanfaatkan fungsi di numpy

#mycsv harus diupload agar code bisa dieksekusi

myfile = open("syntstar2555.csv", "r")
data = np.loadtxt(myfile, delimiter=",", skiprows=1) # skip baris pertama (ke-1) pada data
for j in range(len(data)):
  x = data[:, 0]
  y = data[:, 1]
myfile.close()
```

```
#Definisikan titik data yang akan di-interpolasi
#Gunakan finer and regular mesh
n = 10
xfine = np.linspace(x[0], x[len(data) - 1], n*100) #membuat 1000 titik data baru dalam rentang dat
```

```
# interpolasi dengan fungsi linear (p=1)
print("Nilai dari data yang didapat pada bintang di file syntstar2555")
y1 = scipy.interpolate.interp1d(x, y, kind = 'linear')
y1(xfine)
```

```
#Input data dapat ditanam di kode python, atau interaktif, atau baca dari file.
#Berikut ini merupakan contoh dari pembacaan file worksheet dalam format file teks CSV
#yang berisi dua kolom data dengan memanfaatkan fungsi di numpy

#mycsv harus diupload agar code bisa dieksekusi

myfile2 = open("star255_rev.csv", "r")
data2 = np.loadtxt(myfile2, delimiter=",", skiprows=1) # skip baris pertama (ke-1) pada data
for j in range(len(data2)):
  x2 = data2[:, 0]
  y2 = data2[:, 1]
myfile2.close()
```

```
#Definisikan titik data yang akan di-interpolasi
#Gunakan finer and regular mesh
n = 10
xfine2 = np.linspace(x2[0], x2[len(data2) - 1], n*100) #membuat 1000 titik data baru dalam rentang dat
```

```
# interpolasi dengan fungsi linear (p=1)
print("Nilai dari data yang didapat pada bintang di file star255_rev")
ya = scipy.interpolate.interp1d(x2, y2, kind = 'linear')
ya(xfine2)
```

```
# Plot data dan hasil interpolasi (y0, y1, y2, dan y3)
# pylab.plot(x, y, 'o', label = 'data point') 
pylab.plot(xfine, y1(xfine), label = 'spektrum model')
pylab.plot(xfine2, ya(xfine2), label = 'spektrum bintang yang diamati')
pylab.legend()
pylab.xlabel('panjang gelombang')
pylab.ylabel('fluks')
#pylab.savefig(’interpolate.pdf’)
pylab.show()
```

```
# mencari fluks minimum pada data syntstar2555

ymin = min(y)

# mencari panjang gelombang pada fluks minimum

for j in range (len(data)):
  if y[j] == ymin:
    xmin = x[j]
print("Panjang gelombang diamnya adalah " + str(xmin))
```

```
myfile2 = open("star255_rev.csv", "r")
data2 = np.loadtxt(myfile2, delimiter=",", skiprows=1) # skip baris pertama (ke-1) pada data
for j in range(len(data2)):
  x2 = data2[:, 0]
  y2 = data2[:, 1]

# mencari fluks minimum pada data star255_rev

ymin2 = min(y2)

print(ymin2)

# mencari panjang gelombang pada fluks minimum

for j in range (len(data2)):
  if y2[j] == ymin2:
    xmin2 = x2[j]

print("Panjang gelombangnya adalah " + str(xmin2))
```

```
# mencari perbedaan panjang gelombang

dL = xmin2 - xmin
print("Perbedaan panjang gelombangnya adalah " + str(dL))
```

```
# mencari nilai Vr(kecepatan radial)

c = 2.9989*10**8
Vr = (dL/xmin)*c

print("Maka panjang kecepatan radial Vr adalah "+str(Vr))
```
