```
#Regresi Linear

import numpy as np
from scipy.optimize import curve_fit
import pylab
import pandas as pd

def f(x, a1, a0):
  return a1*x + a0

data = pd.read_csv('dataaskom.csv')
x = (data['varX'])
y = (data['varY'])

init_vals = [1, 1]
popt, pcov = curve_fit(f, x, y, p0=init_vals)
print('Best parameter values: {}'.format(popt))

xfitted = np.linspace(x[0], x[len(data)-1], 50)
pylab.plot(x, y, 'o', label = 'data $(x_i,y_i)$')
pylab.plot(xfitted, f(xfitted, *popt), '-', label = 'fit $f(x_i)$')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.legend()
pylab.show()
```

```
# Regresi kuadratik

import numpy as np
from scipy.optimize import curve_fit
import pylab
import pandas as pd

def f(x, a2, a1, a0):
  return a2*(x**2) + a1*x + a0 

data = pd.read_csv('dataaskom.csv')
x = (data['varX'])
y = (data['varY'])

init_vals = [1, 1, 1 ]
popt, pcov = curve_fit(f, x, y, p0=init_vals)
print('Best parameter values: {}'.format(popt))

xfitted = np.linspace(x[0], x[len(data)-1], 50)
pylab.plot(x, y, 'o', label = 'data $(x_i,y_i)$')
pylab.plot(xfitted, f(xfitted, *popt), '-', label = 'fit $f(x_i)$')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.legend()
pylab.show()
```

```
# Regresi kubik

import numpy as np
from scipy.optimize import curve_fit
import pylab
import pandas as pd

def f(x, a3, a2, a1, a0):
  return a3*(x**3) + a2*(x**2) + a1*x  + a0 

data = pd.read_csv('dataaskom.csv')
x = (data['varX'])
y = (data['varY'])

init_vals = [1, 1, 1, 1 ]
popt, pcov = curve_fit(f, x, y, p0=init_vals)
print('Best parameter values: {}'.format(popt))

xfitted = np.linspace(x[0], x[len(data)-1], 50)
pylab.plot(x, y, 'o', label = 'data $(x_i,y_i)$')
pylab.plot(xfitted, f(xfitted, *popt), '-', label = 'fit $f(x_i)$')
pylab.xlabel('x')
pylab.ylabel('y')
pylab.legend()
pylab.show()
```
