```
#Regresi Linear

import numpy as np
from scipy.optimize import curve_fit
import pylab
import pandas as pd

def f(x, a1, a0):
  return a1*x + a0

data = pd.read_csv('dataaskom2.csv')
data = data.sort_values(by=['Teff'],ascending=True)
x = np.array(data['Teff'])
y = np.array(data['logN_Be'])
x = x[:68]
y = y[:68]

init_vals = [1, 1]
popt, pcov = curve_fit(f, x, y, p0=init_vals)
print('Best parameter values: {}'.format(popt))

xfitted = np.linspace(x[0], x[len(data)-2], 50)
pylab.plot(x, y, 'o', label = 'data $(x_i,y_i)$')
pylab.plot(xfitted, f(xfitted, *popt), '-', label = 'fit $f(x_i)$')
pylab.xlabel('Teff')
pylab.ylabel('Log N(Be)')
pylab.legend()
pylab.show()
```
